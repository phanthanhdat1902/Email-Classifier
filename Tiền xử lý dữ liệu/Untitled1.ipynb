```python
import regex as re
import os
from text_preprocessing import preprocess_text
from text_preprocessing import (
    to_lower,
    remove_email, 
    remove_url, 
    remove_punctuation, 
    lemmatize_word, 
    remove_number, 
    stem_word, 
    remove_stopword,
    remove_whitespace
)
import mailparser

### Lay duong dan den tat ca cac file trong folder
def read_files_name(folderPath):
    result = []
    for dirPath, dirs, files in os.walk(folderPath):
        for fileName in files:
            fName = os.path.join(dirPath, fileName)
            result.append(fName)
    return result


def read_forders_Path():
    easyHam = read_files_name('dataset\easy_ham')
    hardHam = read_files_name('dataset\hard_ham')
    spam = read_files_name('dataset\spam')
    return easyHam, hardHam, spam


def strip_url(mess):
    return re.sub(r'http[s]?://\S+|www\.\S+', ' ', mess)


def strip_html(mess):
    # clean html_tag
    clean = re.compile('<.*?>')
    mess = re.sub(clean, ' ', mess)
    # clean html_space_white
    return re.sub('&nbsp;', ' ', mess)



def strip_email_header(message):
    message = mailparser.parse_from_string(message)
    subject = message.subject
    body = message.body

    return '\n'.join((subject, body))




def preprocessing_data(mess):
    # processing_function_list = [
    #     str.lower,
    #     strip_url,
    #     remove_url,
    #     strip_html,
    #     remove_email,
    #     remove_number,
    #     remove_stopword,
    #     remove_punctuation,
    #     remove_whitespace

    # ]

    preprocess_functions = [strip_email_header, to_lower, remove_email, remove_url,strip_url,strip_html, remove_number, remove_punctuation, remove_stopword, lemmatize_word]
    aftermess = preprocess_text(mess, preprocess_functions)

    # return ' '.join(aftermess)
    return aftermess

def handler():
    a,b,c = read_forders_Path()
    with open(a[0], 'r', encoding='latin') as fi:
        print(preprocessing_data(fi.read()))




if __name__ == "__main__":
    handler()
```
