```python
f = open('../Data/english', 'r+')
stopwords = f.read()
stopwords=stopwords.split("\n")
f.close()
```

```python
stopwords
```

```python
def splitRDD(x):
    temp=[]
    temp=x[0].split(" ")
    for i in range(0,len(stopwords)):
        try :
            index=temp.index(stopwords[i])
            temp.pop(index)
        except:
            continue
    return (temp,x[1])
```

```python
rdd1=rdd.map(lambda x : splitRDD(x))
```
